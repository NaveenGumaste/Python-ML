```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random

random.seed(42)
num_points = 35
x = np.linspace(0, 10, num_points)
y = 2 * x + 1 + np.random.normal(0, 1, num_points)

model = LinearRegression().fit(x.reshape(-1, 1), y)
predicted_y = model.predict(x.reshape(-1, 1))

data = pd.DataFrame({'x': x, 'y': y, 'predicted_y': predicted_y})
data.to_csv('linear_regression_data.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5])  # Independent variable (feature)
y = np.array([2, 4, 5, 4, 5])  # Dependent variable (target)

# Calculate mean of X and y
mean_X = np.mean(X)
mean_y = np.mean(y)

# Calculate the slope (coefficient)
numerator = np.sum((X - mean_X) * (y - mean_y))
denominator = np.sum((X - mean_X) ** 2)
slope = numerator / denominator

# Calculate the intercept
intercept = mean_y - slope * mean_X

# Predict using the linear model
y_pred = slope * X + intercept

# Print the coefficients
print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")

# Plot the data and regression line
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred, color='red', linewidth=2, label="Regression Line")
plt.xlabel("Independent Variable (Feature)")
plt.ylabel("Dependent Variable (Target)")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("./linearregression-data2.csv")
print(data)
```

```python
x=np.array(data[['Roll NO']])
y=np.array(data[['Total']])
print(x)
print(y)
plt.scatter(x,y)
plt.xlabel("ROLL NO")
plt.ylabel("TOTAL")
```

```python
alpha=0.01
iterations=500
b1=0 #weight1
b0=0 #weight0

for i in range (iterations):
    y_pred=b1*x+b0
    error = y_pred-y

    b1 = b1 - (2*alpha/len(y))*np.sum(error*x)
    b0 = b0 - (2*alpha/len(y))*np.sum(error)

y_predict=b1*x+b0
plt.scatter(x,y)
plt.plot(x,y_predict)
```
